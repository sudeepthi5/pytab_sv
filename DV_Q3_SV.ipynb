Using COVID data, a combination of column chart and line graph was drawn for moving average and the number of daily cases for Hillsborough.

#### importing the file

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r'C:\Users\HP\Downloads\Combined-1.xlsx', sheet_name='Hillsborough', names = ['Date', 'Daily Number', 'Moving Average']) 

df.head()
```

```python

s_df = df.plot('Date','Daily Number', color='red')

df.plot('Date', 'Moving Average', kind='bar', xticks=[], width = 1, ax = s_df)
```
